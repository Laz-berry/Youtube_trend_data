```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("KR_youtube_trending_data.csv")
```

```python
df.head(10)
```

```python
df.info()
```

```python
# 같은 날짜의 중복된 영상을 찾아서 제거, 고유한 video_id를 subset으로 설정, 
# 마지막 값을 남김(좋아요, 코멘트가 많음)
df['trending_date'].value_counts()

# 약 3700개의 중복 데이터 존재
df = df.drop_duplicates(subset=['trending_date', 'video_id'], keep='last')
```

```python
# 2020-08-20 ~ 2023-03-23까지는 946일
len(df['trending_date'].unique())
```

```python
# 채널id, 상세설명(정제하기 어려운 데이터) 제거

df = df.drop(['channelId', 'description', 'thumbnail_link'], axis=1)
```

```python
df.info()
```

```python
df.head()
```

```python
# tags가 없으면 제목을 tag로 변경
df['tags'] = np.where(df['tags'] == '[None]', df['title'], df['tags'])
```

```python
# json --> CategotyId를 Categoryname으로 변경
df1 = pd.read_json('KR_category_id.json')
df1
```

```python
# categories를 만들기 위한 array
categories = []

for i, item in enumerate(df1['items']):           
    category = item['snippet']['title']
    categories.append({'categoryId': item['id'], 'category_name': category})

# DataFram생성
df_categories = pd.DataFrame(categories)

df_categories
```

```python
# DataFrame merge를 위한 int형변환
df_categories['categoryId'] = df_categories['categoryId'].astype(int)
df_categories.info()
```

```python
df = pd.merge(df, df_categories, how='left')
df
```

```python
# category_name이 없는 Id는 29번
df[df['category_name'].isna()].categoryId.unique()
```

```python
# 읽기 쉽게 이름 변환
df.rename({ 'category_name' : 'category',
          'publishedAt' : 'published_at',
          'channelTitle' : 'channel_title',
          }, axis=1, inplace=True)
```

```python
# tag가 달리지 않은 데이터 24997개
df['tags'].value_counts()
```

```python
df.to_csv('Trend_data.csv', index=0)
```
