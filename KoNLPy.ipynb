```python
import numpy as np
import pandas as pd
import re
import networkx as nx

from konlpy.tag import Okt
from collections import Counter
from tqdm import tqdm_notebook, tqdm
```

```python
df = pd.read_csv('Trend_data.csv')
```

```python
df.head(10)
```

```python
len(df)
```

```python
okt = Okt()

noun = []
count = []
n_list = []


# noun에 명사 분석을 대입
for i in tqdm(range(len(df))):
    noun.append(okt.nouns(df['tags'][i]))
    
for j in tqdm(range(len(df))):
    count.append(Counter(noun[j]))
    
for k in tqdm(range(len(df))): 
    n_list.append(count[k].most_common(3))
    
```

```python
for i in range(0, 10):
    locals()['n_list'+str(i)] = n_list[i*20000:(i*20000)+20000]
```

```python
with open('test01.txt', 'w', encoding='UTF-8') as f:
    for i in n_list:
        f.write(str(i))
```
